### Ejercicio 1
La sumatoria $1 + 1/2 + 1/4 +1/8 ... $ tal que el error absoluto $e_{abs} < 10^{-1}$.

Esta es una suma infinita dada por: $ \sum\limits_{n=0}^\infty (\frac{1}{2})^n$

La suma total nos da: S = 2

El error absoluto debe ser menor que $10^{-1} = 0.1$, dado esto planteamos la siguiente inecuacion:

y para calcular el $e_{abs}$ utilizaremos: $e_{abs} = |p - p^*|$, ademas utilizaremos 6 cifras significativas.

```python
def encontrar_iteracion_por_error(valor_real, tolerancia):
    suma_parcial = 0.0
    n = 0
    while abs(valor_real - suma_parcial) >= tolerancia:
        suma_parcial += (0.5) ** n
        n += 1
    
    error_final = abs(valor_real - suma_parcial)
    return n, suma_parcial, error_final

# --- Ejecución ---
if __name__ == "__main__":
    p_real = 2.0
    error_deseado = 0.1

    iteraciones, suma_final, error = encontrar_iteracion_por_error(p_real, error_deseado)

    print("Criterio cumplido después de" ,iteraciones, "iteraciones.")
    print("Suma parcial final:", suma_final)
    print("Error absoluto final:" ,error, "que es < error_deseado")
```

Con esto tomaremos a $p = 2$ y $p^*$ = $\sum\limits_{n=0}^\infty (\frac{1}{2})^n$

1. **Primer término:**  
    $S_1 = 1$  
    Error absoluto: $|2 - 1| = 1$ 

2. **Segundo término:**  
    $S_2 = 1 + 0.5 = 1.5$  
    Error absoluto: $|2 - 1.5| = 0.5$ 

3. **Tercer término:**  
    $S_3 = 1.5 + 0.25 = 1.75$  
    Error absoluto: $|2 - 1.75| = 0.25$ 

4. **Cuarto término:**  
    $S_4 = 1.75 + 0.125 = 1.875$  
    Error absoluto: $|2 - 1.875| = 0.125$ 

5. **Quinto término:**  
    $S_5 = 1.875 + 0.0625 = 1.9375$  
    Error absoluto: $|2 - 1.9375| = 0.0625$ 

En la quinta iteracion se cumple el criterio de $e_{abs} < 10^{-1}$.

## Ejercicio 2 (Bubble sort)


### Corrida de escritorio
$v_1=[3, 2, 5, 8, 4, 1]$

| i | Vector |
| -- | -- |
| 0 | $ [3, 2, 5, 8, 4, 1] $ |
| 1 | $ [3, 2, 5, 4, 1, 8] $|
| 2 | $ [3, 2, 4, 1, 5, 8] $|
| 3 | $ [3, 2, 1, 4, 5, 8] $|
...

Resultado final:

$v_{sorted} = [1,2,3,4,5,8]$

Casos de prueba:
* $v_2=[-1, 0, 4, 5, 6, 7]$
* $v_3$ 100_000 número aleatorios entre -200 y 145. 

```python
import random

def bubble_sort(arr):
    n = len(arr)
    cont = 0
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            cont += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True

        if not swapped:
            break
    return arr, cont

v2 = [-1, 0, 4, 5, 6, 7]
sorted_v2, cont_v2 = bubble_sort(v2)
print("v2_sorted:", sorted_v2) 
print("Comparaciones: ", cont_v2)

v3 = [random.randint(-200, 145) for _ in range(100000)]
sorted_v3, cont_v3 = bubble_sort(v3)
print("v3_sorted:", sorted_v3)
print("Comparaciones: ", cont_v3)
```

```python
def fibonacci(n):
    if n < 0:
        raise ValueError("n debe ser un entero no negativo.")
    if n == 0:
        return 0
    else:
        x = 0  
        y = 1  
        for _ in range(1, n):
            z = x + y
            x = y
            y = z
            
        return y

#n = 11
n_11 = 11
resultado_11 = fibonacci(n_11)
print(f"Para n = {n_11}, fib(n) = {resultado_11}")

#n = 84
n_84 = 84
resultado_84 = fibonacci(n_84)
print(f"Para n = {n_84}, fib(n) = {resultado_84}")

#n = 1531
n_1531 = 1531
resultado_1531 = fibonacci(n_1531)
print(f"Para n = {n_1531}, fib(n) = {resultado_1531}")
```

## Graficar!
* El valor de la serie $fib(n)$
* El valor del cociente 

    $\phi \rightarrow \frac{fib(n)} {fib(n-1)} \approx 1.618$ número áureo.


| n | $ fib(n) /fib(n-1) $ |
| -- | -- |
| 2 | $1/1=1 $ |
| 3 | $2/1 = 2$ |
| 4 | $3/2 = 1.5$ |
| 5 | $5/3= 1.66667$ |
| 6 | $8/5= 1.6$ |
| 7 | $13/8 = 1.625 $ |
| 8 | $21/13 = 1.615 $ |
| ... | ... |
|$\infty $ | $ \frac{1 + \sqrt{5}} {2} \approx 1.618$ (número áureo) |

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

def fibonacci(n):
    if n < 0:
        raise ValueError("n debe ser un entero no negativo.")
    if n == 0:
        return 0
    x, y = 0, 1
    for _ in range(1, n):
        x, y = y, x + y
    return y

def graficar_analisis_fibonacci(n_max=20):
    n_valores = list(range(1, n_max + 1))
    fib_valores = [fibonacci(n) for n in n_valores]
    n_cociente = list(range(2, n_max + 1))
    cociente_valores = [fib_valores[i] / fib_valores[i-1] for i in range(1, len(fib_valores))]
    phi = (1 + np.sqrt(5)) / 2

    fig = make_subplots(
        rows=2, cols=1,
        subplot_titles=(
            "Crecimiento de la Serie de Fibonacci",
            "Convergencia al Número Áureo (φ)"
        )
    )

    fig.add_trace(go.Scatter(
        x=n_valores, 
        y=fib_valores, 
        mode='lines+markers',  
        name='fib(n)',
        line=dict(dash='dash') 
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=n_cociente, 
        y=cociente_valores, 
        mode='lines+markers', 
        name='Cociente',
        line=dict(dash='dot') 
    ), row=2, col=1)
    

    fig.add_hline(y=phi, line_dash="dash", line_color="red", annotation_text=f"Número Áureo φ ≈ {phi:.4f}", row=2, col=1)

    fig.update_layout(
        height=800,
        title_text="Análisis Gráfico de la Serie de Fibonacci",
        template="plotly_dark",
        showlegend=False
    )
    fig.update_yaxes(title_text="Valor de fib(n)", row=1, col=1)
    fig.update_yaxes(title_text="Cociente fib(n)/fib(n-1)", row=2, col=1)
    fig.update_xaxes(title_text="n", row=1, col=1)
    fig.update_xaxes(title_text="n", row=2, col=1)
    
    fig.show()

if __name__ == "__main__":
    graficar_analisis_fibonacci(n_max=20)
```

## Ejercicio 4


```python
import plotly.graph_objects as go

def serie_geometrica(n):
    suma_total = 0.0
    for i in range(1, n + 1):
        suma_total += 1 / i
    return suma_total

terminos_x = list(range(100, 1001))
sumas_y = [serie_geometrica(k) for k in terminos_x]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=terminos_x,
    y=sumas_y,
    mode='lines',
    name='Suma Parcial', 
    line=dict(color='cyan', width=2)
))

fig.add_trace(go.Scatter(
    x=[100, 1000],  
    y=[sumas_y[0], sumas_y[-1]], 
    mode='markers+text', 
    name='Puntos Clave',
    marker=dict(color='red', size=10, symbol='diamond'),
    text=[f"Suma={sumas_y[0]:.2f}", f"Suma={sumas_y[-1]:.2f}"],
    textposition="top center" 
))

fig.update_layout(
    title='Comportamiento de la Serie Armónica con Puntos Clave',
    xaxis_title='Número de Términos (n)',
    yaxis_title='Valor de la Suma',
    template='plotly_dark',
    height=600,
    showlegend=True 
)

fig.show()
```
