## Taller 03 Series de Taylor y Polinomios de Lagrange

```python
import matplotlib.pyplot as plt
import numpy as np
```

Con $ln(x), x_0 = 1, n = 3$

$f(x) = ln(x)$

#### Y sus derivadas

$f'(x) = x^{-1}$ , $f''(x) = -x^{-2}$ y $f'''(x) = 2x^{-3}$

$P(x) = \frac{1}{3} (x-1)^{3} - \frac{(x-2)^{2}}{2} + (x+1)$

```python
# Grafico de polinomio de primer orden: (x-1)

x = range(-3, 5)
y = [xi - 1 for xi in x]

plt.plot(x, y, label='P(x) = x - 1', color='blue')
plt.title('Polinomio de Taylor de Primer Orden de ln(x)')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
# Grafico de polinomio de segundo orden: -((x-2)^2)/2 + (x-1)

x = np.linspace(-1, 6, 100 )
y = [-((xi - 2)**2) / 2 + (xi - 1) for xi in x]
plt.plot(x, y, label='P(x) = -((x-2)^2)/2 + (x+1)', color='orange')
plt.title('Polinomio de Segundo Orden')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
# Grafico de polinomio de tercer orden : (1/3)(x-1)^3 - ((x-2)^2)/2 + (x-1)
x = np.linspace(-1, 6, 100)
y = [(1/3)*(xi - 1)**3 - ((xi - 2)**2)/2 + (xi - 1) for xi in x]
plt.plot(x, y, label='P(x) = (1/3)(x-1)^3 - ((x-2)^2)/2 + (x-1)', color='green')
plt.title('Polinomio de Tercer Orden')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
# Corte de las tres aproximaciones en un mismo grafico con f(x) = ln(x)
x = np.linspace(0.1, 4, 100)

# Función original ln(x)
y = np.log(x)

# Polinomio de primer orden: P1(x) = x - 1
y1 = x - 1

# Polinomio de segundo orden: P2(x) = -((x-2)^2)/2 + (x-1)
y2 = -((x - 2)**2) / 2 + (x - 1)

# Polinomio de tercer orden: P3(x) = (1/3)(x-1)^3 - ((x-2)^2)/2 + (x-1)
y3 = (1/3)*(x - 1)**3 - ((x - 2)**2)/2 + (x - 1)

# Graficar todas las funciones
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = ln(x)', color='red', linewidth=2)
plt.plot(x, y1, label='P₁(x) = x - 1', color='blue', linestyle='--')
plt.plot(x, y2, label='P₂(x) = -((x-2)²)/2 + (x-1)', color='orange', linestyle='--')
plt.plot(x, y3, label='P₃(x) = (1/3)(x-1)³ - ((x-2)²)/2 + (x-1)', color='green', linestyle='--')

# Marcar el punto de intersección en x = 1
plt.plot(1, 0, 'ko', markersize=8, label='Punto x₀ = 1')

plt.title('Comparación de Aproximaciones de Taylor para ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(1, color='purple', linewidth=0.5, ls=':', alpha=0.7)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(loc='best')
plt.ylim(-3, 2)
plt.show()
```

## Polinomios de Lagrange

#### Ejercicio 1: Polinomio de Lagrange

**Puntos dados:** $(0,0), (30,0.5), (60,3\sqrt{2}), (90,1)$

$$P(x) = 0 \cdot L_0(x) + 0.5 \cdot L_1(x) + 3\sqrt{2} \cdot L_2(x) + 1 \cdot L_3(x)$$

Donde:

$$L_0(x) = \frac{(x-30)(x-60)(x-90)}{(0-30)(0-60)(0-90)}$$

$$L_1(x) = \frac{(x-0)(x-60)(x-90)}{(30-0)(30-60)(30-90)}$$

$$L_2(x) = \frac{(x-0)(x-30)(x-90)}{(60-0)(60-30)(60-90)}$$

$$L_3(x) = \frac{(x-0)(x-30)(x-60)}{(90-0)(90-30)(90-60)}$$

```python
def polinomio_lagrange(x, x_points, y_points):
    n = len(x_points)
    result = 0
    
    for i in range(n):
        # Calcular el término L_i(x)
        term = y_points[i]
        for j in range(n):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term
    
    return result

x_points_1 = np.array([0, 30, 60, 90])
y_points_1 = np.array([0, 0.5, 3*np.sqrt(2), 1])

x_vals = np.linspace(-10, 100, 300)
y_vals = [polinomio_lagrange(x, x_points_1, y_points_1) for x in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='Polinomio de Lagrange')
plt.plot(x_points_1, y_points_1, 'ro', markersize=10, label='Puntos dados', zorder=5)

# Añadir líneas punteadas desde los puntos a los ejes
for i in range(len(x_points_1)):
    plt.plot([x_points_1[i], x_points_1[i]], [0, y_points_1[i]], 'g--', alpha=0.3, linewidth=1)
    plt.plot([0, x_points_1[i]], [y_points_1[i], y_points_1[i]], 'g--', alpha=0.3, linewidth=1)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de Lagrange - Ejercicio 1')
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()

print("Ejercicio 1 - Verificación:")
print(f"Puntos dados: {list(zip(x_points_1, y_points_1))}")
print(f"\nEl polinomio pasa por los puntos:")
for i in range(len(x_points_1)):
    y_calc = polinomio_lagrange(x_points_1[i], x_points_1, y_points_1)
    print(f"P({x_points_1[i]}) = {y_calc:.6f}, esperado: {y_points_1[i]:.6f}")
```

#### Ejercicio 3: Polinomio de Lagrange

Puntos dados: $(-2,5), (1,7), (3,11), (7,34)$

$$P(x) = 5 \cdot L_0(x) + 7 \cdot L_1(x) + 11 \cdot L_2(x) + 34 \cdot L_3(x)$$

Donde:

$$L_0(x) = \frac{(x-1)(x-3)(x-7)}{(-2-1)(-2-3)(-2-7)}$$

$$L_1(x) = \frac{(x+2)(x-3)(x-7)}{(1+2)(1-3)(1-7)}$$

$$L_3(x) = \frac{(x+2)(x-1)(x-3)}{(7+2)(7-1)(7-3)}$$
